```python
import numpy as np
import math

def grav_acc(s_x, s_y, planet_mass):
    '''
    takes in x,y components of position vector from the planet and mass of planet
    finds acceleration components
    
    input: s_x, s_y, planet_mass
    output: a_x (m/s), a_y (m/s)
    '''
    s = math.sqrt(s_x**2 + s_y**2)
    #formula for a is GMp/s^2
    gravityConstant = 6.67e-11
    a = (gravityConstant*planet_mass)/(s**2)

    sin_beta = -(s_x)/s
    cos_beta = -(s_y)/s

    a_x = a * sin_beta
    a_y = a * cos_beta

    return a_x, a_y

#==========================================================================================
sx = np.array([0, -4787 * 10**3, 6771 * 10**3])
sy = np.array([6371 * 10**3, -4787 * 10**3, 0])
#ax = grav_acc(0, 6371 * 10**3, 5.972 * 10**24)

for i in range(len(sx)):
    ax, ay = grav_acc(sx[i], sy[i], 5.9722 * 10**24)
    mag = (sx[i]**2  + sy[i]**2)**(1/2)
    print(f"x acceleration {np.round(ax, 2)}, y acceleration {np.round(ay, 2)}", 
          f"altitude {mag - (6371 * 10**3)}")
```
